# Linked List

## What is linked list?

* Linked list is a form of sequential collection and it does not have to be in order. LL is made up of independent nodes that may contain any type of data and has a reference to the next node in the LL.
* LL must be traversed sequentially.
* The nodes of a LL are not contiguous, they are not in continuous memory locations.
* The 'head' node is a must as it gives a reference to start accessing the LL in memory.
* Each node has data and reference to the next node < physical location to next node>
* The 'tail' node allows for insertion of element at the end of LinkedList without having to traverse through the entire LL everytime { O(n) }

## Difference b/w arrays and linked lists.

* Each element/node of a LL is an independent object.
* In Arrays the cells are not independent units, so they can not be blatantly deleted, even if the value in the cell of an array is deleted, the cell still remains in memory.
* Size of linked list is not predefined. They can be varied at any time in memory. An array size needs to be specified beforehand.
* Insertion/Removal of values in arrays are computationally expensive, but in Linked List they are very efficient.
* Accesing an element is much more efficient in Array since the elements are indexed, and the index of the element is known, hence O(1). But in LL, the access has to be done sequentially from head towards tail.

## Different types of Linked List.

* Singly LL
* Circular Singly LL.
* Doubly LL.
* Circular Doubly LL.

* Singly Linked List- Each node has data and reference < physical address > to next node in the LL. Last node reference is null.


* Circular Singly Linked List- Last node is referenced to the first node.Last node has reference to first node. {circularly linked}

* Doubly Linked List- Each node has references to both the node ahead of it and behind it, so bidirectional traversal is allowed. We can go ahead of back.

* Circularly Double Linked List- In addition to allowing bi-directional movement, the last node is also linked to the first, bidirectionally.

## Linked Lists in memory.

* They are not located contiguously, but rather randomly in memory and addition of an extra node is also random.
* Allows us dynamic resizing at run-time.
* Traversing has to be sequential since memory location is random.

## Insertion of values in a Linked List.

* Beginning of Linked List.
* After a node in the middle of Linked List.
* End of Linked List.

* If assigning new value at begining, assign memory location in heap. Give reference of this new value to Head and a reference from this new value to the previous first value. Break reference between head and older first value.

* In the middle of LL, start traversing till insertion point is reached, assign new memory location, draw reference from previous node. Break the existing reference from previous node to the older next node, and then allot another reference from the newly inserted value to the next node.

* At the end, traverse to the end, allocate memory in heap for the new element, update the reference of the previously last node to the new last node and assign reference of the new last node to the tail of the LL. O(n) time complexity, because complete traversal of the LL is required.

## Insertion algorithm in Single Linked List.

* Insertion also works in a manner more or less similar to that of deletion of node in a LL. 
* First we check if the LL exists, if it doesn't just add the new node as the only node.
* If it is to be inserted at the beginning of the LL, the head is referenced to the next node and the newNode is given the self.head
* If it is to be inserted at the end of the LL, the next reference of the node is assigned None, the previous tail next reference is given to newNode and the self.tail is given to newNode.
* For insertion in middle just check the procedure to delete from middle.

````python
class Node:
  def __init__(self, value=None):
    self.value = value  #value of node.
    self.next= None     #Reference of next node.
class SingleLinkedList:
  def __init__(self):
    self.head= None  #Head
    self.tail=None   #Tail 
  def __iter__(self): # function to print the linked list to compare original and updated node values
    node= self.head
    while node:
      yield node
      node= node.next
  def insertSLL(self, value, location):
    newNode=  Node(value)
    if self.head is None: # if the SLL is empty we're referencing both the head and tail to newly created node
      self.head= newNode
      self.tail = newNode
    else:
      if location == 0: #inserting element at beginning of SLL
        newNode.next= self.head
        self.head= newNode
      elif location == -1 : # insert element at end of SLL
        newNode.next= None
        self.tail.next= newNode
        self.tail= newNode
      else:
        tempNode= self.head
        index=0
        while index< location-1:
          tempNode= tempNode.next
          index+=1
        nextNode= tempNode.next
        tempNode.next= newNode
        newNode.next= nextNode
        if tempNode == self.tail:
          self.tail = newNode
````

## Delete entire Linked List.
