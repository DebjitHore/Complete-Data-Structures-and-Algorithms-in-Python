## Creation of a `Circular Single Linked List`

* In Circular Singly Linked List the tail node has reference to the head node. That is the only difference

```python
class Node:
	def __init__(self, value=None):
		self.value=value
		self.next=None
class CircularSinglyLinkedList:
	def __init__(self):
		self.head= None
		self.tail= None
	def __iter__(self):
		node= self.head
		while node:
			yield node
			if node.next==self.head:
				break
			node=node.next

#Function to create a Circular Single Linked List

	def CreateCSLL(self, nodeValue):
		node=Node(nodeValue)
		node.next=node
		self.head= node 
		self.tail= node 
		return "The CSLL has been created"



circularSLL= CircularSinglyLinkedList()
circularSLL.CreateCSLL(nodeValue=55)

print([node.value for node in circularSLL])
```

* Time complexity of creation of a circular single linked list is O(1)
* Space complexity is also O(1)
